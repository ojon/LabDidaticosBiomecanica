#Vamos calcular os ângulos do membro inferior no plano sagital

```python
import numpy as np
from IPython.display import display, Latex
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter, filtfilt
from IPython.display import HTML, display
%matplotlib inline
```

```python
filename = 'dam_lin_un.csv'
data = np.loadtxt(filename, skiprows=2, unpack=False)
markers = ['ACR', 'HIP', 'KNEE', 'FIBULA', 'ANKLE', 'HEEL', 'MT5', 'TOE']
```

coluna 0 frames, coluna 1 tempo, a partir deste colunas com pares X Y para os marcadores.

Vamos filtrar os dados com um filtro butterworth passa baixa

```python
freqAmostragem = 30.0
freqNiquist = freqAmostragem/2;
freqCorte = 5;
ordem = 4

t = np.arange(0,1,.01);
# Butterworth filter
b, a = butter(ordem, (freqCorte/freqNiquist), btype = 'low')

rawData = np.copy(data);

#tempo e frames não precisam ser filtrados
tempo = rawData[:, 0];
frame = rawData[:, 1];
```

```python
#separando em variáveis para ficar mais didático
ACR    =  np.column_stack((filtfilt(b, a,(rawData[:, 2])), filtfilt(b, a,rawData[:,3])))
HIP    =  np.column_stack((filtfilt(b, a,(rawData[:, 4])), filtfilt(b, a,rawData[:,5])))
KNEE   =  np.column_stack((filtfilt(b, a,(rawData[:, 6])), filtfilt(b, a,rawData[:,7])))
FIBULA =  np.column_stack((filtfilt(b, a,(rawData[:, 8])), filtfilt(b, a,rawData[:,9])))
ANKLE  =  np.column_stack((filtfilt(b, a,(rawData[:, 10])), filtfilt(b, a,rawData[:,11])))
HEEL   =  np.column_stack((filtfilt(b, a,(rawData[:, 12])), filtfilt(b, a,rawData[:,13])))
MT5    =  np.column_stack((filtfilt(b, a,(rawData[:, 14])), filtfilt(b, a,rawData[:,15])))
TOE    =  np.column_stack((filtfilt(b, a,(rawData[:, 16])), filtfilt(b, a,rawData[:,17])))
```

###Primeiro os Angulos dos segmentos

```python
#tronco 
tetha10 = 180 -np.rad2deg(np.arctan2(ACR[:,1] - HIP[:,1], ACR[:,0] - HIP[:,0]))

#coxa
tetha21 = 180 - np.rad2deg(np.arctan2(HIP[:,1] - KNEE[:,1], HIP[:,0] - KNEE[:,0]))

#perna
tetha43 = 180 - np.rad2deg(np.arctan2(FIBULA[:,1] - ANKLE[:,1], FIBULA[:,0] - ANKLE[:,0]))

#meioPe
tetha65 = 180 - np.rad2deg(np.arctan2(HEEL[:,1] - MT5[:,1], HEEL[:,0] - MT5[:,0]))

#antePe
tetha76 = 180 - np.rad2deg(np.arctan2(MT5[:,1] - TOE[:,1], MT5[:,0] - TOE[:,0]))
```

```python
plt.clf()
plt.figure(figsize=(9,12))
ax1 = plt.subplot(5,1,1)
plt.plot(gaitPercent, tetha10)
plt.title('Tronco')
plt.grid()

plt.subplot(5,1,2, sharex=ax1, sharey=ax1)
plt.plot(gaitPercent, tetha21)
plt.title('Coxa')
plt.grid()

plt.subplot(5,1,3, sharex=ax1, sharey=ax1)
plt.plot(gaitPercent, tetha43)
plt.title('Perna')
plt.ylabel('graus')
plt.grid()

plt.subplot(5,1,4, sharex=ax1, sharey=ax1)
plt.plot(gaitPercent, tetha65)
plt.title('MeioPe')
plt.grid()

plt.subplot(5,1,5, sharex=ax1,  sharey=ax1)
plt.plot(gaitPercent, tetha76)
plt.title('Antepe')
plt.xlabel('% ciclo da passada')
plt.grid()
```

##Agora os ângulos das articulações

```python
tethaHip = -tetha10 + tetha21 
tethaKnee = tetha21 - tetha43
tethaAnkle = (tetha43 - tetha65 + 90)*-1
tethaMT_PH = -(tetha65 - tetha76)
```

```python
plt.clf()
plt.figure(figsize=(9,12))
ax1 = plt.subplot(4,1,1)
plt.title('Quadril')
plt.plot(gaitPercent, tethaHip)
plt.ylim(-20,25)
plt.ylabel('graus')
plt.plot([0, 100],[0, 0], '--r')
plt.grid()

plt.subplot(4,1,2, sharex=ax1)
plt.plot(gaitPercent, tethaKnee)
plt.title('Joelho')
plt.plot([0, 100],[0, 0], '--r')
plt.grid()

plt.subplot(4,1,3, sharex=ax1)
plt.plot(gaitPercent, tethaAnkle)
plt.title('Tornozelo')
plt.plot([0, 100],[0, 0], '--r')
plt.ylim(-15,25)
plt.grid()

plt.subplot(4,1,4, sharex=ax1)
plt.plot(gaitPercent, tethaMT_PH)
plt.title('Metatarso')
plt.plot([0, 100],[0, 0], '--r')
plt.xlabel('% ciclo da passada')
plt.ylim(-15,50)
plt.grid()
```

###Velocidades 

```python
#calculo dos intervalos de tempos
timeDiff = data[1:,1] - data[:-1,1]

#velocidades angulares
omegaHip = (tethaHip[1:] - tethaHip[:-1])/timeDiff
omegaKnee = (tethaKnee[1:] - tethaKnee[:-1])/timeDiff
omegaAnkle = (tethaAnkle[1:] - tethaAnkle[:-1])/timeDiff
omegaMT_PH = (tethaMT_PH[1:] - tethaMT_PH[:-1])/timeDiff
```

```python
plt.clf()
plt.figure(figsize=(9,12))
ax1 = plt.subplot(4,1,1)
plt.title('Quadril')
plt.plot(gaitPercent[1:], omegaHip)
plt.ylabel('graus/s')
plt.plot([0, 100],[0, 0], '--r')
plt.grid()

plt.subplot(4,1,2, sharex=ax1)
plt.plot(gaitPercent[1:], omegaKnee)
plt.title('Joelho')
plt.plot([0, 100],[0, 0], '--r')
plt.grid()

plt.subplot(4,1,3, sharex=ax1)
plt.plot(gaitPercent[1:], omegaAnkle)
plt.title('Tornozelo')
plt.plot([0, 100],[0, 0], '--r')
plt.grid()

plt.subplot(4,1,4, sharex=ax1)
plt.plot(gaitPercent[1:], omegaMT_PH)
plt.title('Metatarso')
plt.plot([0, 100],[0, 0], '--r')
plt.grid()
plt.xlabel('% ciclo da passada')
```

###Aceleração

```python
#calculo dos intervalos de tempos
timeDiff = data[1:,1] - data[:-1,1]


#aceleraçôes angulares
alfaHip = (omegaHip[1:] - omegaHip[:-1])/timeDiff[1:]
alfaKnee = (omegaKnee[1:] - omegaKnee[:-1])/timeDiff[1:]
alfaAnkle = (omegaAnkle[1:] - omegaAnkle[:-1])/timeDiff[1:]
alfaMTPH = (omegaMT_PH[1:] - omegaMT_PH[:-1])/timeDiff[1:]
```
