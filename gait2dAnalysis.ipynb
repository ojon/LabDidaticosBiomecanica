#Vamos calcular os ângulos do membro inferior no plano sagital

```python
import numpy as np
from IPython.display import display, Latex
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter, filtfilt
from IPython.display import HTML, display
%matplotlib inline
```

```python
filename = 'data/gaitDataFromSkillSpector.csv'
data = np.loadtxt(filename, skiprows=2, unpack=False)
markers = ['ACR', 'HIP', 'KNEE', 'FIBULA', 'ANKLE', 'HEEL', 'MT5', 'TOE']
```

Vamos filtrar os dados com um filtro butterworth passa baixa

```python
freq = 25.0
t = np.arange(0,1,.01);
# Butterworth filter
b, a = butter(4, (5/(freq/2)), btype = 'low')

rawData = np.copy(data);

#coisa feia preciso descobrir como fazer uma iteracao decente neste objeto ....
# filter with phase shift correction
data[:, 1] = filtfilt(b, a, rawData[:, 1])
data[:, 2] = filtfilt(b, a, rawData[:, 2])
data[:, 3] = filtfilt(b, a, rawData[:, 3])
data[:, 4] = filtfilt(b, a, rawData[:, 4])
data[:, 5] = filtfilt(b, a, rawData[:, 5])
data[:, 6] = filtfilt(b, a, rawData[:, 6])
data[:, 7] = filtfilt(b, a, rawData[:, 7])
data[:, 8] = filtfilt(b, a, rawData[:, 8])
data[:, 9] = filtfilt(b, a, rawData[:, 9])
data[:, 10] = filtfilt(b, a, rawData[:, 10])
data[:, 11] = filtfilt(b, a, rawData[:, 11])
data[:, 12] = filtfilt(b, a, rawData[:, 12])
data[:, 13] = filtfilt(b, a, rawData[:, 13])
data[:, 14] = filtfilt(b, a, rawData[:, 14])
data[:, 15] = filtfilt(b, a, rawData[:, 15])
data[:, 16] = filtfilt(b, a, rawData[:, 16])
data[:, 17] = filtfilt(b, a, rawData[:, 17])
```

```python
## shift de 2 colunas e calculo das diferenças das matrizes
segsCoord = data[:, 2:16] - data[:, 4:18]
```

###Primeiro os Angulos dos segmentos

```python
#quadril
tetha10 = np.rad2deg(np.unwrap(np.arctan2(segsCoord[:,1],segsCoord[:,0])))

#coxa
tetha21 = np.rad2deg(np.unwrap(np.arctan2(segsCoord[:,3],segsCoord[:,2])))

tetha32 = np.rad2deg(np.unwrap(np.arctan2(segsCoord[:,5],segsCoord[:,4])))

#perna
tetha43 = np.rad2deg(np.unwrap(np.arctan2(segsCoord[:,7],segsCoord[:,6])))

tetha54 = np.rad2deg(np.unwrap(np.arctan2(segsCoord[:,9],segsCoord[:,8])))

#meioPe
tetha65 = np.rad2deg(np.unwrap(np.arctan2(segsCoord[:,11],segsCoord[:,10])))

#antePe
tetha76 = np.rad2deg(np.unwrap(np.arctan2(segsCoord[:,13],segsCoord[:,12])))
```

```python
plt.clf()
plt.figure(figsize=(9,12))
ax1 = plt.subplot(4,1,1)
plt.title('Angulos dos Segmentos')
plt.plot(data[:,1], tetha21)
plt.ylabel('Coxa')
plt.grid()

plt.subplot(4,1,2, sharex=ax1)
plt.plot(data[:,1], tetha43)
plt.ylabel('Perna')
plt.grid()

plt.subplot(4,1,3, sharex=ax1)
plt.plot(data[:,1], tetha65)
plt.ylabel('MeioPe')
plt.grid()

plt.subplot(4,1,4, sharex=ax1)
plt.plot(data[:,1], tetha76)
plt.ylabel('Antepe')
plt.xlabel('tempo(s)')
plt.grid()
```

```python
plt.clf()
plt.figure(figsize=(9,8))
for i in range(0,np.size(data,0),2):
    plt.plot(data[i,2::2],data[i,3::2],'-bo')
```

##Agora os ângulos das articulações

```python
tethaHip = tetha10 - tetha21
tethaKnee = tetha21 - tetha43
tethaAnkle = tetha43 - tetha65 + 90
tethaMT_MP = tetha65 - tetha76
```

```python
plt.clf()
plt.figure(figsize=(9,12))
ax1 = plt.subplot(3,1,1)
plt.title('Angulos das Articulacoes')
plt.plot(data[:,1], tethaHip)
plt.ylabel('Quadril')
plt.grid()

plt.subplot(3,1,2, sharex=ax1)
plt.plot(data[:,1], tethaKnee)
plt.ylabel('Joelho')
plt.grid()

plt.subplot(3,1,3, sharex=ax1)
plt.plot(data[:,1], tethaAnkle)
plt.ylabel('Tornozelo')
plt.grid()
```

###Velocidades 

```python
#calculo dos intervalos de tempos
timeDiff = data[1:,1] - data[:-1,1]

#velocidades angulares
omegaHip = (tethaHip[1:] - tethaHip[:-1])/timeDiff
omegaKnee = (tethaKnee[1:] - tethaKnee[:-1])/timeDiff
omegaAnkle = (tethaAnkle[1:] - tethaAnkle[:-1])/timeDiff
```

```python
plt.clf()
plt.figure(figsize=(9,12))
ax1 = plt.subplot(3,1,1)
plt.title('Velocidade Angulares das Articulacoes')
plt.plot(data[1:,1], omegaHip)
plt.ylabel('Quadril')
plt.grid()

plt.subplot(3,1,2, sharex=ax1)
plt.plot(data[1:,1], omegaKnee)
plt.ylabel('Joelho')
plt.grid()

plt.subplot(3,1,3, sharex=ax1)
plt.plot(data[1:,1], omegaAnkle)
plt.ylabel('Tornozelo')
plt.grid()
```

Melhorou bastante agora com o filtro

###Aceleração

```python
#calculo dos intervalos de tempos
timeDiff = data[1:,1] - data[:-1,1]


#aceleraçôes angulares
alfaHip = (omegaHip[1:] - omegaHip[:-1])/timeDiff[1:]
alfaKnee = (omegaKnee[1:] - omegaKnee[:-1])/timeDiff[1:]
alfaAnkle = (omegaAnkle[1:] - omegaAnkle[:-1])/timeDiff[1:]
```

```python
plt.clf()
plt.figure(figsize=(9,12))
ax1 = plt.subplot(3,1,1)
plt.title('Aceleracoes Angulares das Articulacoes')
plt.plot(data[2:,1], alfaHip)
plt.ylabel('Quadril')
plt.grid()

plt.subplot(3,1,2, sharex=ax1)
plt.plot(data[2:,1], alfaKnee)
plt.ylabel('Joelho')
plt.grid()

plt.subplot(3,1,3, sharex=ax1)
plt.plot(data[2:,1], alfaAnkle)
plt.ylabel('Tornozelo')
plt.grid()
```
